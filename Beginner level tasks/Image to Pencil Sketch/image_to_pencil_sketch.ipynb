# **IMAGE TO PENCIL SKETCH**

```python
#importing the required libraries
import numpy as np
import cv2 as cv #used for computer vision tasks
from google.colab.patches import cv2_imshow
```

```python
#loading the image
img = cv.imread('/content/dog.jpg')
print(type(img))
print(img.shape)
```

```python
#original image
cv2_imshow(img)
```

**CONVERTING THE ORIGINAL IMAGE INTO GREY IMAGE**

```python
img_grey=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
#gray image
cv2_imshow(img_grey)
```

**INVERTED THE GRAY SCALE IMAGE**

```python
inverted_image = 255 - img_grey
```

```python
#inverted the gray scale image
cv2_imshow(inverted_image)
```

**APPLY GUASSIAN BLUR**

```python
blurred = cv.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
#blur image
cv2_imshow(blurred)
```

```python
inverted_blurred = 255 - blurred
```

```python
cv2_imshow(inverted_blurred)
```

**FINAL PENCIL SKETCH **

```python
pencil_sketch = cv.divide(img_grey, inverted_blurred, scale=256.0)
```

```python
#final pencil sketch
cv2_imshow(pencil_sketch)
```

**ORIGINAL IMAGE VS PENCIL SKETCH IMAGE**

```python
cv2_imshow(img)

cv2_imshow(pencil_sketch)
```
